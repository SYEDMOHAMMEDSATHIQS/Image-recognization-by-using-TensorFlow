```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Loading the MNIST dataset into training and testing sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Printing the shape of the dataset to understand its dimensions
print(f"Training set shape (images): {x_train.shape}")
print(f"Testing set shape (images): {x_test.shape}")
```

```python
# Visualizing the first 10 training images along with their labels
plt.figure(figsize=(10, 5))
for i in range(10):  # Loop through the first 10 images
    plt.subplot(2, 5, i + 1)  # Create a 2x5 grid for displaying images
    plt.imshow(x_train[i], cmap='gray')  # Display the image in grayscale
    plt.title(f"Label: {y_train[i]}")  # Show the corresponding label
    plt.axis('off') #Turn off axes for better visualization

plt.show()  # Display the figure
```

```python
# Visualizing the first 10 testing images along with their labels
plt.figure(figsize=(10, 5))
for i in range(10):  # Loop through the first 10 images
    plt.subplot(2, 5, i + 1)  # Create a 2x5 grid for displaying images
    plt.imshow(x_test[i], cmap='gray')  # Display the image in grayscale
    plt.title(f"Label: {y_test[i]}")  # Show the corresponding label
    plt.axis('off') #Turn off axes for better visualization

plt.show()  # Display the figure
```

```python
from tensorflow.keras.datasets import cifar10

# Loading CIFAR-10 into training and testing sets
(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) =cifar10.load_data()

# x_train_cifar: Training images (50,000 examples of 32x32x3 pixels)
# y_train_cifar: Labels for training images (0-9, representing  categories)
print(f"Training set shape (images): {x_train_cifar.shape}")
print(f"Testing set shape (images): {x_test_cifar.shape}")
```

```python
# We have designed the class names for the CIFAR-10 categories - labelling
class_names = [
 'Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
 'Dog', 'Frog', 'Horse', 'Ship', 'Truck'
]
```
